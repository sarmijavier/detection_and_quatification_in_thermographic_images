```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image from file
image = cv2.imread('image1.png')

# Make a copy to keep the original image unchanged
original_image = image.copy()

# Convert the image to grayscale (HoughCircles works on grayscale images)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise
# This improves edge detection and circle detection
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Detect circles using the Hough Circle Transform
# dp=1  resolution of the accumulator equals the image resolution
# minDist=40  minimum distance between detected circle centers
# param1=40  upper threshold for the Canny edge detector
# param2=45  threshold for center detection (higher = stricter)
# minRadius and maxRadius  limits for detected circle sizes
circles = cv2.HoughCircles(
    blurred, cv2.HOUGH_GRADIENT, dp=1, minDist=40,
    param1=40, param2=45, minRadius=55, maxRadius=145
)

# List to store only circles that are fully inside the image boundaries
complete_circles = []

# Get the height (h) and width (w) of the blurred image
h, w = blurred.shape

# Loop through each detected circle
for x, y, r in circles[0]:
    # Keep the circle only if it is completely inside the image
    if (x - r > 0) and (x + r < w) and (y - r > 0) and (y + r < h):
        complete_circles.append((x, y, r))

# Copy the original image to draw the detected circles on it
output = image.copy()

# Draw each complete circle on the output image
for x, y, r in complete_circles:
    x, y, r = int(x), int(y), int(r)
    cv2.circle(output, (x, y), r, (5, 5, 5), 8)  # Draw the circle outline
    cv2.circle(output, (x, y), 2, (1, 1, 1), 5)  # Draw the center point

# Print how many complete circles were found
print("Circles found:", len(complete_circles))

# Convert BGR image (OpenCV format) to RGB for displaying with matplotlib
output_rgb = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)

# Save the final image in BGR format (OpenCV uses BGR)
cv2.imwrite("results/detected_circles_1.png", output)

# Show the final image with detected circles
plt.figure(figsize=(8, 8))
plt.imshow(output_rgb)
plt.show()
```
