```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image1.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (9, 9), 0)

clahe = cv2.createCLAHE(clipLimit=3.0)
blurred = clahe.apply(blur)

circles = cv2.HoughCircles(
    blurred, cv2.HOUGH_GRADIENT, dp=1, minDist=60, param1=60, param2=60, minRadius=100, maxRadius=145
)
complete_circles = []
h, w = blurred.shape

for x, y, r in circles[0]:
    if (x-r > 0) and (x+r < w) and (y-r > 0) and (y+r < h):
        complete_circles.append((x, y, r))

output = img.copy()

for x, y, r in complete_circles:
    x, y, r = int(x), int(y), int(r)
    cv2.circle(output, (x, y), r, (5,5,5), 3)
    cv2.circle(output, (x, y), 2, (1,1,1), 3)
    
print("Circles found:", len(complete_circles))
output_rgb = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8, 8))
plt.imshow(output_rgb)
plt.show()
```
